```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
from sklearn.preprocessing import PolynomialFeatures
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (15,12)
```

### Cargar datos

```python
df = pd.read_excel("Tarea Ajustar Modelo.xlsx")
df.shape
```

# MODELO POLINOMIO 2

```python
x = df["X"].to_numpy()
y = df["Y"].to_numpy()
```

```python
x= x[:,np.newaxis]
y= y[:,np.newaxis]
```

```python
inds = x.ravel().argsort()      
x = x.ravel()[inds].reshape(-1,1)
y = y[inds] 
```

```python
from sklearn.preprocessing import PolynomialFeatures
polynomial_features= PolynomialFeatures(degree=2)
xp = polynomial_features.fit_transform(x)
xp.shape
```

```python
import statsmodels.api as sm

model = sm.OLS(y, xp).fit()
ypred = model.predict(xp) 
```

```python
plt.scatter(x,y)
plt.plot(x,ypred)
plt.title("Polin√≥mio de Grado 2", size = 26)
```

```python
model.summary()
```

## MODELO POLINOMIO 3

```python
polynomial_features3= PolynomialFeatures(degree=3)
xp3 = polynomial_features3.fit_transform(x)
xp3.shape

model3 = sm.OLS(y, xp3).fit()
ypred3 = model3.predict(xp3) 

plt.scatter(x,y)
plt.plot(x,ypred3)
plt.title("Polin√≥mio de Grado 3", size = 26)
```

```python
model3.summary()
```
