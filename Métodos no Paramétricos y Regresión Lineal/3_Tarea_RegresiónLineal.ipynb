### a) Utilice los datos de las variables X,Y para ajustar un modelo lineal simple
### b) Apóyese en los encabezados de columnas para efectuar los cálculos complementarios.

```python
# COLUMNA X^2
df["X^2"] = df["X"]**2

#CÁLCULO DE LA MEDIA DE X Y DE Y
media_x = np.mean(df["X"])
media_y = np.mean(df["Y"])

#COLUMNA x
df["x"] = df["X"] - media_x

#COLUMNA y
df["y"] = df["Y"] - media_y

#COLUMNA xy
df["xy"] = df["x"] * df["y"]

#COLUMNA x^2
df["x^2"] = df["x"]**2

df.head(10)
```

### c) Obtenga estimaciones púntuales de los parámetros : $\hat{\alpha}$, $\hat{\beta}$, s2, sus varianzas y errores estándares.

```python
beta = df["xy"].sum() / df["x^2"].sum()
alfa = media_y - beta * media_x
print("El valor de beta es:",beta,"\n","El valor de alfa es:",alfa)
```

```python
#COLUMNA Y EST
df["Y_est"] = alfa + df["X"]*beta
df["e"] = df["Y"] - df["Y_est"]
df["e^2"] = df["e"]**2
df["y^2"] = df["y"]**2 
df.head(10)
```
